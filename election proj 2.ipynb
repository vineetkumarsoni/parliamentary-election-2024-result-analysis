```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results_2024_winners.csv')
```

```python
df.head()
```

```python
winning_party_by_state = pd.crosstab(df['State'], df['Winning Party'])
print(winning_party_by_state)
```

```python
# Assuming df is your dataframe
# Clean the 'Margin Votes' column by removing non-numeric values
df['Margin Votes'] = pd.to_numeric(df['Margin Votes'], errors='coerce')

# Create a distribution plot using displot
sns.displot(df['Margin Votes'], kde=True)  # kde=True adds a kernel density estimate
```

```python
total_votes = df['Margin Votes'].sum()

print("Total Votes in 2024 Lok Sabha Election:", total_votes)
```

```python
state_wise_win_percentage = pd.crosstab(df['State'], df['Winning Party'], normalize='index')
print(state_wise_win_percentage)
```

```python
#Pie chart of winning party distribution

party_counts = df['Winning Party'].value_counts()
party_counts.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#Bar chart for top 10 margin votes

top_10_margin = df.nlargest(10, 'Margin Votes')
top_10_margin.plot(kind='bar', x='PC Name', y='Margin Votes')
plt.show()
```

```python
#winning party and runner-up party comparison

party_comparison = df[['Winning Party', 'Runner-up Party']]
pd.crosstab(party_comparison['Winning Party'], party_comparison['Runner-up Party'])
```

```python
#State-wise total votes cast

state_wise_total_votes = df.groupby('State').sum()
print(state_wise_total_votes)
```

```python
#Creating a pivot table for winning party and runner-up party comparison

party_pivot = pd.pivot_table(df, index='Winning Party', columns='Runner-up Party', values='Margin Votes', aggfunc=np.mean)
print(party_pivot)
```
